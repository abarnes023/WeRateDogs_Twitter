## WeRateDogs Act Report

We all love cute pictures of dogs with a little humor mixed in, so it's not surprising the WeRateDogs Twitter account is so popular. Are there any trends we can see that may have caused this Twitter account to be so popular as opposed to a similar account? Some of the account's tweets are also more popular than others (more retweets and favorites), so what is it that causes some tweets to get retweeted and favorited so often? These are the questions that this report tries to answer, and it led to a few insights.

First, WeRateDogs tweeted far more often in the first few months of the account's creation compared to each month since. In November 2015 there were almost 300 tweets, and in December 2015 well over 350 tweets were sent out. Compare that with the amount of tweets per month from April 2016 onward, a period in which WeRateDogs did not reach even 100 tweets in any month.

Interestingly, the inclusion of a dog name or dog stage ('puppo', 'floofer', 'pupper', 'doggo') did not correlate with the popularity of a tweet. A dog's rating did not either. When a tweet is favorited by a lot of people though, it is also retweeted by a lot of people.

The clearest trend can be seen below, and it is true for both retweets and favorites. WeRateDogs's tweets grow in popularity (measured here by number of retweets and favorites) as time goes on. This makes a lot of sense if we assume that the Twitter account has gradually gained followers over time, leading to them having a bigger audience seeing each tweet.

WeRateDogs has clearly been doing something right, with over 7 million followers to date. The multitude of tweets immediately following the account's creation may very well have helped them build a solid foundation of followers right away. The retweets and favorites gained from these followers would help lead to more followers, and it may have just continued recurrently from there. Intuitively, we can assume that more followers leads to more retweets and favorites. So the best path to Twitter popularity may be to tweet, tweet, tweet right away, and then patiently wait for your followers to grow.
